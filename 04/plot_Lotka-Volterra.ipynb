# ロトカ-ボルテラ モデルのプロット

## 準備

```python
# inline表示のため
%matplotlib inline
```

```python
# パッケージの読み込み
import numpy as np
import matplotlib.pyplot as plt
```

```python
# プロットのための関数定義
def plotLotkaVolterra2sp(data, title = "Lotka-Volterra model", sp1label="Sp1", sp2label="Sp2", figuresize = (14, 6)):
    """
    この関数は，ロトカ-ボルテラモデルのシミュレーション結果をプロットします．

    Parameters
    ----------
    data : array
        シミュレーションの結果．
    title : str
        プロットのタイトル．
    sp1label : str
        種1の凡例のラベル．
    sp2label : str
        種2の凡例のラベル．
    figuresize: tuple
        図のサイズを指定．
    """
    plt.figure(figsize=figuresize)
    p1 = plt.plot(data[:,0], data[:,1], marker="o", markeredgewidth=0)
    p2 = plt.plot(data[:,0], data[:,2], marker="o", markeredgewidth=0)
    plt.title(title)
    plt.legend((p1[0], p2[0]), (sp1label, sp2label), loc=2)
    plt.xlabel("time t")
    plt.ylabel("population x, y")
    plt.grid(True)
    plt.show()
```

## 被食-捕食系
$$
\begin{cases}
  \frac{dx}{dt} = (a-by)x\\
  \frac{dy}{dt} = (cx-d)y
\end{cases}
$$

```python
# データの読み込み
data_prey_predator = np.loadtxt("./output_4-6.csv",delimiter=",")
```

```python
data_prey_predator
```

```python
# データのプロット
plotLotkaVolterra2sp(data_prey_predator, "Prey-Predator", "Prey", "Predator", (14,6))
```

`plotLotkaVolterra2sp`関数の引数は`title`，`sp1label`，`sp2label`には日本語を使わない方が良い．  
デフォルトのままだと日本語用フォントが設定されていないため“豆腐”のような記号に置き換えられてしまう．

## 競争系
$$
\begin{cases}
  \frac{dx}{dt} = (a- bx -cy)x\\
  \frac{dy}{dt} = (d - ex -fy)y
\end{cases}
$$

```python
data_competitive = np.loadtxt("./output_4-7.csv",delimiter=",")
```

```python
plotLotkaVolterra2sp(data_competitive, "Competitive", "Sp1", "Sp2", (14,6))
```

## 共生系
$$
\begin{cases}
  \frac{dx}{dt} = (a- bx +cy)x\\
  \frac{dy}{dt} = (d + ex -fy)y
\end{cases}
$$

```python
data_mutualism = np.loadtxt("./output_4-8.csv",delimiter=",")
```

```python
plotLotkaVolterra2sp(data_mutualism, "Mutualism", "Sp1", "Sp2", (14,6))
```
